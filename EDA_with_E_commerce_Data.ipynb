```python
#distibution of users by country 
a=df['country'].value_counts().head(10)
a.plot.bar()
```

```python
#distibution of users by gender
b=df['gender'].value_counts()
b.plot.bar()
```

```python
#distibution of users by gender
c=df['language'].value_counts()
c.plot.bar()
```

```python
#filter data where value products Sold > 1 or not zero
df1=df[df['productsSold']>1]
df1
```

```python
plt.figure(figsize=(25,12))
sns.barplot(x='country', y='productsSold', hue='gender', data=df1.head(36)).set(title='Country Performance (products Sold) by Gender')
```

```python
plt.figure(figsize=(25,12))
sns.barplot(x='country', y='productsSold', hue='language', data=df1.tail(36)).set(title='Country Performance (products Sold) by Language')
```

```python
#total product sold by country
a=df.groupby(by=['country'], as_index=False)['productsSold'].sum()
a
```

```python
plt.subplots(figsize=(25,12))
plt.plot(a['country'].head(18), a['productsSold'].head(18))
plt.xlabel('country')
plt.ylabel('products Sold')
plt.title('number of products Sold for country')
plt.show()
```

```python
#TOP 5 most country where is product sold
a.sort_values(by='productsSold', ascending=False).head()
```

```python
#total product sold by language
df1=df.groupby(by=['language'], as_index=False)['productsSold'].sum()
df1.sort_values(by='productsSold', ascending=False)
```
